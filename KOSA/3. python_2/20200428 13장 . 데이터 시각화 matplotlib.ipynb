#### 2) subplots() 함수로 서브플롯 집합 추가

```python
x = np.arange(0,10,0.01)
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
x = np.arange(0,10,0.01)
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].plot(x,np.sin(x)); axes[0,1].plot(x,np.cos(x))
axes[1,0].plot(x,np.tanh(x)); axes[1,1].plot(x,np.sin(x)*np.cos(x))
```

```python
x = np.arange(0,7,0.01)
```

```python
def sin_cos(x):
    return np.sin(x)*np.cos(x)
```

```python
func_list = [np.sin, np.cos, np.tanh,sin_cos]
```

```python
fig,axes = plt.subplots(nrows =2, ncols =2)
for i ,ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x))
    
plt.show()
# 이런식으로 for구문을 돌려서 할 수 있는데, 이떄는 flat 이 필요핟. 
```

```python
# subplots 함수의 인수를 어떻게 선언하느냐에 따라 그래프 영역이 ㅏ뉨. 
# ncols : 4개의 열로 나눠라 라는 뜻 
fig,axes = plt.subplots(ncols =4) 
for i ,ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x))
    
plt.show()
```

```python
# subplots 함수의 인수를 어떻게 선언하느냐에 따라 그래프 영역이 ㅏ뉨. 
# nrows : 4개의 행으로 나눠라 라는 뜻 
fig,axes = plt.subplots(nrows =4) 
for i ,ax in enumerate(axes.flat):
    ax.plot(x,func_list[i](x))
    
plt.show()
```

#### 2) plot() 
- plot() 함수는 주어진 x,y 값을 선과 점으로 표시해줌 

```python
fig,axes = plt.subplots(2,2,figsize = (8,5)) 
fig.suptitle("figure sample plots")
axes[0,0].plot([1,2,3,4],'ro-') # 빨간(r), 동그라미(o),실선(-)
                                # x만 주면 y는 자동으로 같은 값을 가짐
axes[0,1].plot(np.random.randn(4,10),np.random.randn(4,10),'cs-.')
                # cyan(c) , 네모(s), 대시닷선(-.)
axes[1,0].plot(np.linspace(0,5),np.cos(np.linspace(0,5))) # 0~5까지 50개
axes[1,1].plot([3,6],[3,5],'b^:')# 파랑(b), 세모(^), 점선(:)
axes[1,1].plot([4,5],[5,4],'kx--') # 검정(k), x(X), 대시선(--)
plt.show()
```

#### 4) 축과 눈금

```python
fig,axes = plt.subplots(2,2,figsize=(8,5))
plt.subplots_adjust(hspace=0.4,wspace=0.3)
axes[0,0].scatter(x,y,marker=".")
axes[0,0].set(title="An Example Axes",ylabel ="Y-Axis",xlabel="X-Axis")
            # 그래프의 제목과 축의 제목, 눈금, 눈금 레이블 등을 설정
axes[0,1].scatter(x,y,marker="^",c="r")
axes[0,1].set_xlim(0,5) # x축의 범위 설정 
axes[0,1].set_ylim(-2,2) # y축의 범위 설정 
axes[0,1].set_xlabel("X 0~5") # x축 레이블(이름)을 설정 
axes[0,1].set_ylabel("Y -2~2")  # y축 레이블(이름)을 설정 

axes[1,0].scatter(x,y,marker=".",c="c")
axes[1,0].set(xticks=range(1,8,2), # 눈금의 위치를 리스트 형식으로 지정 
             xticklabels=[3,100,-12,"foo"], # 눈금의 레이블을 리스트로 지정 
             yticks=[-2,0,1,2],
             yticklabels=[-200,0,100,"Max"])

axes[1,1].scatter(x,y,marker="v",c="b")
axes[1,1].set(yticks=[-2,0,1,2],yticklabels=[-200,0,100,"Max"])
axes[1,1].spines['top'].set_visible(False) # 그래프의 박스 경계를 설정 
axes[1,1].spines["bottom"].set_position(("outward",10))
axes[1,1].grid(True) # 눈금선을 보여줌 
plt.show()
```

#### 5) 축 공유

```python
x = np.arange(0,10)
y1 = 0.5* x**2
y2 = -1 * y1
```

```python
fig,ax1 = plt.subplots()
ax1.plot(x,y1,'g^:')
ax1.set_xlabel('x data')
ax1.set_ylabel('y1data',color ='g')
#  하나의 그래프 영역에 다른 축을 지정하고 싶을 때 

ax2= ax1.twinx() # x축을 공유 하고, y 축을 오른쪽으로 표현
ax2.plot(x,y2,'bv--')
ax2.set_ylabel('y2 data',color ="b")

ax3 = ax1.twiny() # y축을 공유하고 x축을 위쪽에 표시해주는 것 
ax3.plot(-x,y1,'ro-.')
ax3.set_xlabel('-x data', color ='r')
plt.show()
```

#### 11) 주기표를 이용한 스타일 지정

```python
plt.style.use("default")
print(plt.rcParams["axes.prop_cycle"].by_key()['color'])
```

```python
from cycler import cycler
my_cycler = (cycler("color",['r','g','b','c','m','y','k'])+
            cycler(linestyle=["-","--",":","-.","-","--",":"])+
            cycler(lw=np.linspace(5,20,7)))
plt.rcParams["axes.prop_cycle"] = my_cycler
plt.plot([1,2],[1,1]); plt.plot([1,2],[2,2]); plt.plot([1,2],[3,3])
plt.plot([1,2],[4,4]); plt.plot([1,2],[5,5]); plt.plot([1,2],[6,6])
plt.plot([1,2],[7,7]); plt.plot([1,2],[8,8]); plt.show()
```
