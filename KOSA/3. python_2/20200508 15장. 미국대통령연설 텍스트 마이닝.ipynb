#### 기본 코드 

```python
import pandas as pd
txt = pd.read_csv("./speech/George Bush Speech - Inaugural address.txt",sep="#",header = None)

from nltk.tokenize import RegexpTokenizer
from nltk.tag import pos_tag
from itertools import chain
import re

def get_nouns_list(line): 
    line = str(line) #요소하나를 문자로 바꾸줌
    ret = RegexpTokenizer("[\w]{2,}") # 정규표현식으로 토크나이즈화 함. 
    tokens = ret.tokenize(line)
    tagged_list =pos_tag(tokens) # 품사를 tagging 함. 
    pattern = re.compile("NN?")
    nouns_list = [t[0] for t in tagged_list if pattern.match(t[1])]
    return nouns_list
```
