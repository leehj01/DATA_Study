```python
sns.set() # style = "darkgrid", palette= "deep" , font="sans-serif",font_scale =1
ax = sns.scatterplot(x= "petal_length", y= "petal_width", hue="species",style = "species",data =iris)

# 이미 되어있는 스타일 기본 설정에 따라, 값에 따라서 설정이 됨. 
# mat 은 점의 갯수만큼 컬러를 지정해줘야하는데 , 예는 간단하게 그려줌.
```

#### 2) 범주형 그래프 : 범주형 데이터를 이용함

```python
# barplot 포인트 추정치와 신뢰 구간을 막대그래프로 표시
# 남자와 여잔의 평균 생존을 알려줌

titanic = sns.load_dataset("titanic")
ax = sns.barplot(x ="sex", y = "survived" ,hue ="class",  data = titanic)
```

```python
# 범주구간의 관측수를 표시 함.
# 타이타닉의 객실 등급별로 구분

ax = sns.countplot(x = "deck", data= titanic, palette="Greens_d")
```

```python
# 도형을 이용해서 점 추정치와 신뢰구간을 표시  pointplot
# 막대기 구간은 신뢰구간을 의미함.
ax = sns.pointplot(x="class",y="survived",hue="sex", data = titanic,
                  palette={"male":"g","female":"m"},
                  markers=["^","o"],linestyles=['-','--'])
```

```python
# boxplot 사분이 그래프 
# 수염 밖에 있는 값은 이상치. 이상치를 찾는것으로 많이 사용

ax = sns.boxplot(x = "alive", y ="age",hue = "adult_male", data = titanic)
```

#### 4) 회귀 그래프 : 평균으로 수렴하는 결과를 가짐 

```python
# lmplot : 데이터를 이용해 선형회귀 모델을 만들고 그래프를 표현

sns.set() # 기존의 설정을 초기화 해주세요 라는 의미 
ax = sns.lmplot(x="sepal_width", y = "sepal_length",data = iris, hue = "species")
```

```python
ax = sns.regplot(x = "petal_length",y= "petal_width",data = iris ) 
# 실제 기울기와 편차는 알려주지 않기 떄문에, 따로 넘파이를 이용해서 구해줘야함 
```

```python
from scipy import stats 
s,i,r,p,se = stats.linregress(x= iris.petal_length,y = iris.petal_width)
print("splope" , s)
print("intercept", i)
print("r_value",r)
print("p_value",p)
print("std_err",se)
```

#### 5) matrix plots : 행렬 그래프

```python
iris = sns.load_dataset("iris")
species = iris.pop("species")

ax = sns.clustermap(iris, figsize=(5,5))
```

```python
# heatmap : 데이터 색으로 인코딩 된 직사각형 행렬로 표시함 
# annot=True : 값을 함께 표시함 
ax = sns.heatmap(iris.corr(),vmin= -1,vmax=1,annot=True, cmap ="YlGnBu")

# https://jiffyclub.github.io/palettable/cubehelix/  내가 원하는 색 
```

```python
import numpy as np
mask = np.zeros_like(iris.corr())
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(iris.corr(),mask=mask, annot=True,
                    linewidths = .5 ,cmap="YlGnBu", square =True)
    
# mask 속성을 이용하면 그래프의 일부분만 그릴 수 있다. 
```

### 3.4. 다중 그래프를 위한 그리드
#### 1) facetgrid

```python
import seaborn as sns
sns.set()
iris = sns.load_dataset("iris")
```

```python
g = sns.FacetGrid(iris, col="species", hue = "species") # 영역을 나눠줌 
# 열을 '종'으로 나눈다는 의미. 
```

```python
g = sns.FacetGrid(iris, col="species", hue = "species")
g.map(plt.hist, "sepal_length")
g.set_axis_labels(y_var="Count") # lable 이름 설정 
```

```python
# scatterplot() 함수를 이용해서 이변량 그래프를 그림. 
# map()함수의 첫번째 인수에 어떤 함수 객체가 들어느냐에 따라 인수의 수와 이름이 달라짐 


g = sns.FacetGrid(iris, col ="species", hue = "species")
g.map(sns.scatterplot, "petal_width", "petal_length",size=iris.sepal_length)
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
g = sns.FacetGrid(titanic, col="survived",row="sex")
g.map(plt.hist,"age")
```

```python
g = sns.PairGrid(iris,hue = "species")
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.regplot)
```
