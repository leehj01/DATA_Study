```python
import pandas as pd
```

```python
redwine = pd.read_csv("winequality-red.csv", sep=";")
```

```python
redwine.head()
```

```python
redwine.loc[redwine.quality==5,:]
```

```python
redwine_X = redwine.iloc[:,:-1]
redwine_y = redwine.iloc[:,-1]
```

```python
import random
random.seed(7902)
for col in range(11):
    redwine_X.iloc[[random.sample(range(len(redwine)), 100)],col] = float('nan')
```

```python
redwine_X.head()
```

```python
redwine_n = pd.concat([redwine_X, redwine_y], axis=1)
```

```python
redwine_n.loc[redwine.quality==5,:]
```

```python
redwine_n.to_csv("이두인_결측치100.csv", encoding="utf-8-sig", index=False)
```

```python
redwine_n = pd.read_csv("이두인_결측치100.csv")
```

```python
import missingno as ms
```

```python
ms.matrix(redwine_n)
```

```python
redwine_mean_by_species = redwine_n.groupby("quality").mean()
```

```python
rmbs = redwine_mean_by_species
rmbs
```

```python
import math
def na2mean(row) :
    for i in range(11):
        if(math.isnan(row[i])) :
            row[i]=rmbs.loc[row["quality"], row.index[i]]
    return row
```

```python
redwine_n2 = redwine_n.apply(na2mean, axis=1)
```

```python
redwine_n2.head()
```

```python
# redwine_n2 = redwine_n.fillna(0)
```

```python
redwine_n2.head(10)
```

```python
X = redwine_n2.iloc[:, :-1]
y = redwine_n2.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3,
                                                    random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(train_X, train_y)
```

```python
rf_model.score(test_X, test_y)
```
